```python
## Objectives:
```

```python
# - Explore match statistics and trends in the ATP tour.
# - Analyze player performance across different surfaces.
# - Identify the most dominant players and matches with the most upsets.
# - Visualize trends in match durations, serve statistics, and more.
```

```python
# 1. Data Overview

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Data

df = pd.read_csv("atp_matches_2000.csv")
df.head()
```

```python
# 2. Data Cleaning
# - Handle missing values
# - Convert dates and categorical values

df['tourney_date'] = pd.to_datetime(df['tourney_date'], format='%Y%m%d')
df.isnull().sum()
```

```python
# 3. Exploratory Data Analysis (EDA)

# Surface Distribution
df['surface'].value_counts().plot(kind='bar', title='Match Surface Distribution')

# Match Duration Statistics
df['minutes'].describe()

# Top 10 Players by Match Wins
top_winners = df['winner_id'].value_counts().head(10)
top_winners.plot(kind='bar', title='Top 10 Players by Match Wins')
```

```python
# 4. Player Performance Analysis

# Serving Statistics Comparison
df['w_1stWon%'] = df['w_1stWon'] / df['w_1stIn']
df['l_1stWon%'] = df['l_1stWon'] / df['l_1stIn']
df[['w_1stWon%', 'l_1stWon%']].mean()
```

```python
# 5. Upset Detection

# Define upsets as cases where a lower-ranked player beats a higher-ranked one
df['upset'] = (df['winner_rank'] > df['loser_rank']) & df['winner_rank'].notnull() & df['loser_rank'].notnull()
df['upset'].value_counts()
```

```python
# 6. Visual Insights

# - Match durations by round

# Boxplot: Match durations by round
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='round', y='minutes')
plt.title('Match Durations by Round')
plt.xlabel('Round')
plt.ylabel('Match Duration (minutes)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# - Win Rate by Surface

# Count wins per player per surface
wins_by_surface = df.groupby(['surface', 'winner_id']).size().reset_index(name='wins')

# Count total matches per player per surface
matches_by_surface = pd.concat([
    df[['surface', 'winner_id']].rename(columns={'winner_id': 'player_id'}),
    df[['surface', 'loser_id']].rename(columns={'loser_id': 'player_id'})
])

total_matches = matches_by_surface.groupby(['surface', 'player_id']).size().reset_index(name='total_matches')

# Merge wins and total matches
merged = pd.merge(wins_by_surface, total_matches, left_on=['surface', 'winner_id'], right_on=['surface', 'player_id'])

# Calculate win rate
merged['win_rate'] = merged['wins'] / merged['total_matches']

# Average win rate by surface
surface_win_rates = merged.groupby('surface')['win_rate'].mean().reset_index()

# Barplot: Average win rate by surface
plt.figure(figsize=(8, 5))
sns.barplot(data=surface_win_rates, x='surface', y='win_rate', hue='surface',  palette='colorblind', legend=False)
plt.title('Average Win Rate by Surface')
plt.ylabel('Win Rate')
plt.xlabel('Surface')
plt.tight_layout()
plt.show()
```
