## Import Package

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

## Load Dataset in 

```python
df = pd.read_csv("netflix_titles.csv")
```

### Preprocess dates

```python
# Preprocess dates
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), format='mixed', errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

### Plot 1: Movies vs TV Shows

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='type')
plt.title('Distribution of Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('plot1_type_distribution.png')
plt.show()
```

### Plot 2: Monthly additions over the years

```python
monthly_additions = df.groupby(['year_added', 'month_added']).size().unstack().fillna(0)
monthly_additions.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Monthly Content Additions Over Years')
plt.xlabel('Year Added')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('plot2_monthly_additions.png')
plt.show()
```

### Plot 3: Release year trends

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['release_year'], bins=40, kde=False)
plt.title('Distribution of Content by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.tight_layout()
plt.savefig('plot3_release_year_trend.png')
plt.show()
```

### Plot 4: Top contributing countries

```python
top_countries = df['country'].dropna().str.split(', ').explode().value_counts().head(10)
plt.figure(figsize=(10, 5))
top_countries.plot(kind='bar')
plt.title('Top 10 Contributing Countries')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('plot4_top_countries.png')
plt.show()
```

### Plot 5: Common genres

```python
top_genres = df['listed_in'].str.split(', ').explode().value_counts().head(10)
plt.figure(figsize=(10, 5))
top_genres.plot(kind='bar', color='teal')
plt.title('Top 10 Common Genres')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('plot5_common_genres.png')
plt.show()
```

### Plot 6: Age ratings comparison

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='rating', hue='type', order=df['rating'].value_counts().index)
plt.title('Age Ratings Comparison for Movies vs TV Shows')
plt.xlabel('Count')
plt.ylabel('Rating')
plt.tight_layout()
plt.savefig('plot6_age_ratings_comparison.png')
plt.show()
```

### Plot 7: WordCloud for descriptions

```python
text = ' '.join(df['description'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('WordCloud of Descriptions')
plt.tight_layout()
plt.savefig('plot7_wordcloud_descriptions.png')
plt.show()
```

### Sacve cleaned to excel for Tablaeu Dashboard

```python
# Load and clean dataset
df = pd.read_csv("netflix_titles.csv")
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), format='mixed', errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['country_clean'] = df['country'].str.split(',').str[0].str.strip()  # Use primary country
df['main_genre'] = df['listed_in'].str.split(',').str[0].str.strip()  # Use primary genre

# Save cleaned dataset to Excel
df.to_excel("netflix_cleaned.xlsx", index=False)
```
