```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('Sport car price.csv')
df.head()
```

```python
# Clean the 'Price (in USD)' column: remove commas and convert to float
df['Price (in USD)'] = df['Price (in USD)'].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
# Group data by 'Year' and calculate the average price
yearly_price_trend = df.groupby('Year')['Price (in USD)'].mean().reset_index()
```

```python
# Plot the average price per year
plt.figure(figsize=(10, 6))
plt.plot(yearly_price_trend['Year'], yearly_price_trend['Price (in USD)'], marker='o')
plt.title('Average Sports Car Price by Year')
plt.xlabel('Year')
plt.ylabel('Average Price (USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Create a scatter plot to visualize the distribution of car prices by year
plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['Price (in USD)'], alpha=0.7)
plt.title('Sports Car Prices by Year')
plt.xlabel('Year')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot both individual prices and average trend
plt.figure(figsize=(10, 6))

# Scatter plot for individual car prices
plt.scatter(df['Year'], df['Price (in USD)'], alpha=0.6, color='dodgerblue', label='Individual Car Prices')

# Line plot for average car prices by year
plt.plot(yearly_price_trend['Year'], yearly_price_trend['Price (in USD)'], color='green', linewidth=2, label='Average Price by Year')

# Add chart title and axis labels
plt.title('Sports Car Prices by Year (Individual and Average)')
plt.xlabel('Year')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save the plot as a PNG image
plt.savefig('sports_car_prices_by_year.png', dpi=300)  # You can change the filename and format

# Show the plot
plt.show()
```
