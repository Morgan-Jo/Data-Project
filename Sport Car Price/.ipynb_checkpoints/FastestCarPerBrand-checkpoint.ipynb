```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from docx import Document
from docx.shared import Inches
```

```python
# Load the dataset
df = pd.read_csv("Sport car price.csv")
```

```python
# show dataset 
df.head()
```

```python
# Convert acceleration to numeric and drop rows with missing values
df['0-60 MPH Time (seconds)'] = pd.to_numeric(df['0-60 MPH Time (seconds)'], errors='coerce')
df_clean = df.dropna(subset=['0-60 MPH Time (seconds)'])
```

```python
# Get the fastest car per brand
fastest_cars = df_clean.loc[df_clean.groupby('Car Make')['0-60 MPH Time (seconds)'].idxmin()]
fastest_cars_sorted = fastest_cars.sort_values('0-60 MPH Time (seconds)')
```

```python
# Create and save the plot
plt.figure(figsize=(12, 8))
sns.barplot(
    data=fastest_cars_sorted,
    x='0-60 MPH Time (seconds)',
    y='Car Make',
    hue='Car Make',
    palette='viridis',
    legend=False
)
plt.title('Fastest Car per Brand (0–60 MPH Time)')
plt.xlabel('0–60 MPH Time (seconds)')
plt.ylabel('Car Make')
plt.tight_layout()
plt.savefig('fastest_car_per_brand.png')
plt.show()
```

```python
# Create a Word document with analysis and plot
doc = Document()
doc.add_heading('Fastest Car per Brand Analysis', 0)

doc.add_paragraph(
    "This analysis identifies the fastest accelerating car (measured by 0–60 MPH time) from each car manufacturer. "
    "Using data grouped by 'Car Make', the car with the lowest 0–60 MPH time in each group was selected. "
    "The resulting cars were sorted and visualized to highlight acceleration performance across brands."
)

doc.add_paragraph("Key Insights:")
doc.add_paragraph("- The analysis uses grouping and filtering techniques in pandas.", style='List Bullet')
doc.add_paragraph("- Fastest car per brand is selected using the minimum 0–60 MPH time.", style='List Bullet')
doc.add_paragraph("- A bar plot visualizes how each brand’s quickest car performs in terms of acceleration.", style='List Bullet')

doc.add_picture('fastest_car_per_brand.png', width=Inches(6.5))
doc.add_paragraph("Figure: Fastest car per brand based on 0–60 MPH time.")

# Save the Word document
doc.save("Fastest_Car_Per_Brand_Analysis.docx")
```
