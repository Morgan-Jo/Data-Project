## Disney+ Data Analysis Project

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# Load dataset
df = pd.read_csv("disney_plus_titles.csv")

# Convert date_added
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Split duration into number and type
df['duration_num'] = df['duration'].str.extract(r'(\d+)').astype(float)
df['duration_type'] = df['duration'].str.extract('([a-zA-Z ]+)$')[0].str.strip()

df.head()
```

# 1. Content Type Distribution

```python
sns.countplot(data=df, x='type', hue='type', palette='pastel', legend=False)
plt.title("Content Type Distribution")
plt.savefig("content_type_distribution.png", bbox_inches='tight')
plt.show()
```

# 2. Monthly Content Additions

```python
monthly_additions = df['date_added'].dt.to_period('M').value_counts().sort_index()
monthly_additions.plot(figsize=(10, 5), title='Monthly Content Additions')
plt.xlabel("Date Added")
plt.ylabel("Number of Titles")
plt.savefig("monthly_content_additions.png", bbox_inches='tight')
plt.show()
```

# 3. Top Countries by Content

```python
top_countries = df['country'].dropna().str.split(', ', expand=True).stack().value_counts().head(10)

df_countries = top_countries.reset_index()
df_countries.columns = ['country', 'count']

sns.barplot(data=df_countries, x='count', y='country', hue='country', legend=False, palette='Set1')
plt.title("Top 10 Countries by Content")
plt.savefig("top_countries_by_content.png", bbox_inches='tight')
plt.show()
```

# 4. Most Common Directors and Cast Members

```python
# Top 10 Directors
top_directors = df['director'].dropna().str.split(', ', expand=True).stack().value_counts().head(10)
sns.barplot(x=top_directors.values, y=top_directors.index, color='green')
plt.title("Top 10 Directors")
plt.savefig("top_10_directors.png", bbox_inches='tight')
plt.show()

# Top 10 Cast Members
top_cast = df['cast'].dropna().str.split(', ', expand=True).stack().value_counts().head(10)
sns.barplot(x=top_cast.values, y=top_cast.index)
plt.title("Top 10 Cast Members")
plt.savefig("top_10_cast.png", bbox_inches='tight')
plt.show()
```

# 5. Genre Frequency Analysis

```python
# Top 15 Genres Bar Chart
genre_counts = df['listed_in'].str.split(', ', expand=True).stack().value_counts().head(15)
sns.barplot(x=genre_counts.values, y=genre_counts.index, color='magenta')
plt.title("Top 15 Genres")
plt.savefig("top_15_genres.png", bbox_inches='tight')
plt.show()

# Word Cloud of Genres
wordcloud_genres = WordCloud(width=800, height=400, background_color='white').generate(' '.join(df['listed_in'].dropna()))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_genres, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of Genres")
plt.savefig("genre_wordcloud.png", bbox_inches='tight')
plt.show()
```

# 6. Content Duration Insights

```python
movie_durations = df[df['type'] == 'Movie']
tv_durations = df[df['type'] == 'TV Show']

sns.histplot(movie_durations['duration_num'].dropna(), bins=20, kde=True)
plt.title("Movie Duration Distribution (minutes)")
plt.savefig("movie_duration_distribution.png", bbox_inches='tight')
plt.show()

sns.countplot(x=tv_durations['duration_num'].dropna())
plt.title("TV Show Season Count Distribution")
plt.savefig("tv_season_distribution.png", bbox_inches='tight')
plt.show()
```

# 7. Age Rating Distribution

```python
sns.countplot(data=df, x='rating', order=df['rating'].value_counts().index, color='orange')
plt.title("Age Rating Distribution")
plt.xticks(rotation=45)
plt.savefig("age_rating_distribution.png", bbox_inches='tight')
plt.show()
```

# 8. Release Year Trends

```python
release_year_counts = df['release_year'].value_counts().sort_index()
release_year_counts.plot(kind='line', figsize=(10, 5), title='Release Year Trends')
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.savefig("release_year-trends.png", bbox_inches='tight')
plt.show()
```

# 9. Text Analysis on Descriptions

```python
all_descriptions = ' '.join(df['description'].dropna())
wordcloud_desc = WordCloud(width=800, height=400, background_color='white').generate(all_descriptions)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_desc, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of Descriptions")
plt.savefig("text_analysis_description.png", bbox_inches='tight')
plt.show()
```

# 10. Genre Comparison: Movies vs TV Shows

```python
genre_split = df[['type', 'listed_in']].dropna()
genre_split = genre_split.assign(genre=genre_split['listed_in'].str.split(', ')).explode('genre')
genre_counts_split = genre_split.groupby(['type', 'genre']).size().reset_index(name='count')

plt.figure(figsize=(12, 6))
sns.barplot(data=genre_counts_split, x='genre', y='count', hue='type')
plt.title("Genre Comparison: Movies vs TV Shows")
plt.xticks(rotation=90)
plt.savefig("genre_comparison.png", bbox_inches='tight')
plt.show()
```
